```python
selected_df=["genres","keywords","tagline","vote_average","director"]
```

```python
for feature in selected_df :
    df[feature]=df[feature].fillna('')
```

```python
comb_df = df["genres"]+' '+ df["keywords"] + ' '+df["tagline"] + ' '  + df['director']
```

```python
comb_df
```

```python
porter_stemmer=PorterStemmer()
```

```python
def stemming(content):
    stemmed_content=re.sub('[^a-zA-Z]', ' ',content)
    stemmed_content = stemmed_content.lower()
    stemmed_content= stemmed_content.split()
    stemmed_content= [porter_stemmer.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content = ' '.join (stemmed_content)
    return stemmed_content
```

```python
comb_df= comb_df.apply(stemming)
```

```python
comb_df
```

```python
#string to numbers
```

```python
vectorizer=TfidfVectorizer()
```

```python
vectorizer.fit(comb_df)
comb_df= vectorizer.transform(comb_df)
```

```python
print(comb_df)
```

```python
similarity = cosine_similarity(comb_df)
```

```python
similarity.shape
```

```python
movie_name=input("Enter The name of movie , u want it to be similar")

```

```python
import difflib
```

```python
list_=df['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_)
print(find_close_match)
```

```python
close_match = find_close_match[0]
index_of_the_movie = df[df.title == close_match]['index'].values[0]


similarity_score=list(enumerate(similarity[index_of_the_movie]))
```

```python
sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

```python

movies_data=df
print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index==index]['title'].values[0]
    if (i<30):
        print(i, '.',title_from_index)
        i+=1

```
